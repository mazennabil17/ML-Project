# preprocessing

## Explore the data

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/weatherHistory.csv')
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

## Fill NAN

```python
data['Precip Type'].ffill(inplace=True)
data.isnull().sum()
```

```python
data.drop(['Loud Cover'], axis=1, inplace=True)

data.head()
```

```python
data.info()
```

## Encoding

```python
data['Formatted Date'] = pd.to_datetime(data['Formatted Date'] ,utc=True)
data['year'] = data['Formatted Date'].dt.year
data['month'] = data['Formatted Date'].dt.month
data['day'] = data['Formatted Date'].dt.day
data['hour'] = data['Formatted Date'].dt.hour
data.drop(['Formatted Date'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()

data['Summary'] = Le.fit_transform(data['Summary'])
data['Daily Summary'] = Le.fit_transform(data['Daily Summary'])
data['Precip Type'] = Le.fit_transform(data['Precip Type'])
data.head()
```

```python
data.info()
```

#Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_features = ['Pressure (millibars)', 'Wind Bearing (degrees)', 'Daily Summary','year']
data[numerical_features] = scaler.fit_transform(data[numerical_features])
```

# Modeling

## Linear regression model

```python
x =data.drop('Temperature (C)', axis=1)
```

```python
x.shape
```

```python
y = data['Temperature (C)']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print("x_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)
print("x_test shape:", x_test.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn import metrics

print('Mean Absolute Error: {:.2f}'.format(metrics.mean_absolute_error(y_test, y_pred)))
print('Mean Squared Error:{:.2f}'.format(metrics.mean_squared_error(y_test, y_pred)))
print('Root Mean Squared Error:{:.2f}'.format(np.sqrt(metrics.mean_squared_error(y_test, y_pred))))
print('R2 Score:{:.2f}'.format(metrics.r2_score(y_test, y_pred)))
```

```python
print('Linear Regression Model:')
print("Train Score {:.2f}".format(model.score(x_train,y_train)))
print("Test Score {:.2f}".format(model.score(x_test, y_test)))
```
